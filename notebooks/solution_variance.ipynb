```python
import os
import sys
# Add parent directory to system path
parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
sys.path.append(parent_dir)

import synthesis.graph_generator as graph_generator
import synthesis.edge_addition_generator as eag
import algorithms.bruteforce as bruteforce
import algorithms.greedy as greedy
import algorithms.ga as ga
import networkx as nx
import numpy as np
```

```python
weighter = graph_generator.UniformWeighter()
#solver = bruteforce.BruteForceOracleSolver()
solver = ga.GeneticSolver()
```

```python
G = nx.watts_strogatz_graph(200, 3, 0.1)
G = weighter.weight_graph(G)
nx.draw(G)
for problem in eag.generate_problems(G, weighter):
    if problem.solution_degree > 1000000:
        continue
    print(problem)
    print('Problem ASPL', problem.aspl)
    sln = solver.solve(problem)
    sln_aspls = [aspl for aspl, _ in sln.trace]
    print('Mean Solution ASPL ', np.mean(sln_aspls))
    print('SD Solution ASPL ', np.std(sln_aspls))
```
